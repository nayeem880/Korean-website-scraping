# Parent Link Extraction

```python
# importing the required modules
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
```

```python
url = "https://class101.net/creative"
base = "https://class101.net"
```

```python
res = requests.get(url)
res
```

```python
soup = BeautifulSoup(res.text, 'html.parser')
print(soup.prettify())
```

```python
tab_links = soup.find_all('a', class_=["NavigationTab__Tab-sc-1sw30ay-1", "fKOHkt"])
tab_links
```

```python
tabs = []
for i in range(1, len(tab_links)-1):
    tabs.append(base+(tab_links[i].attrs['href']))
```

```python
tabs
```

```python
for j  in tabs:
    print(j)
```

```python
df = pd.DataFrame(tabs)
df
```

```python
df.to_csv("Data/parent_links.csv")
```
