```python
# importing the required modules
import requests
from bs4 import BeautifulSoup
import pandas as pd
from selenium import webdriver
import time

df = pd.read_csv('Intermediate_links.csv')
li = list(df["0"])
b_url = "https://m.kin.naver.com"
profile_links = []
item = []


def ps(url):
    print(url)
    driver.get(url)
    time.sleep(3)
    source_code = driver.page_source
    soup = BeautifulSoup(source_code, 'html.parser')
    print('ps')

    # how many times the loop will run
    span = soup.find_all('span', class_ = 'currentPage')
    for a in span:
        min = int(a.next.next.next)
        max = int(a.next.next.next.next.next.next.next.next)
        print("MIN :", min, "MAX :",max)

def start():
    print("start")
    ss = driver.page_source
    soup = BeautifulSoup(ss, 'html.parser') 
    content = soup.find_all('div', class_="content--jBp3G")

    for ii in content:
        a.append(ii.find_all('div', class_=["itemLink--2ljnw"]))

    for j in a:
        for jk in j:
            plinks.append(jk.find_all('a', class_ = "profileLink--16bta"))

    for li in plinks:
        for ji in li:
            if b_url+str(ji.attrs['href']) not in profile_links:
                profile_links.append(b_url+str(ji.attrs['href']))

def click():
    print("click")
    path = u"//a[@class='pageLink linkNext']"
    driver.find_element_by_xpath(path).click()
    
for i in range(len(li)):
    print("Serial:", i)
    min = 0
    max = 0 
# program running here # program running here 
# program running here # program running here 
    driver = webdriver.Chrome(r'C:\Users\USER\chromedriver_win32\chromedriver.exe')
    url = "https://"+li[i]
    plinks = []
    a=[]
    ps(url)
    print("Here")
    
    while(min <= max):
        try:
            print("While. . .")
            start()
            click()
        except:
            break
    driver.quit()
```
